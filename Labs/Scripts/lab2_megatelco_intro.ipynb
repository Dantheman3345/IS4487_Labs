## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
avg_handset_price_by_college = df.groupby('college')['handset_price'].mean()

# Creating the bar chart
avg_handset_price_by_college.plot(kind='bar')
plt.title('Average Handset Price by College Enrollment Status')
plt.xlabel('College (1 = attended, 0 = not attended)')
plt.ylabel('Average Handset Price')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])  # Optional: make labels clearer
plt.show()
```

```python
df['income'].plot(kind='hist', bins=20)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
