```python
import pandas as pd
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

df = pd.read_csv (r'/content/gdrive/MyDrive/Colab Notebooks/adviseinvest (1).csv')
```

```python
# Adjusting the plot scaling and adding count labels
plt.figure(figsize=(10, 6))
mpl.rcParams['font.size'] = 12.0
mpl.rcParams['axes.grid'] = True

# Creating a bar chart with count labels
ax = sns.countplot(x='answered', data=df, palette="Set2")
plt.xlabel('Answered')
plt.ylabel('Count')
plt.title('Relationship between Answered and Number of Accounts')

# Adding count labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')

plt.show()
```

```python

# Load the dataset
df = pd.read_csv(r'/content/gdrive/MyDrive/Colab Notebooks/adviseinvest (1).csv')

# Convert "mobile" into a categorical variable and map values
df['mobile'] = df['mobile'].astype('category')
df['mobile'] = df['mobile'].map({1: 'Provided mobile phone', 0: 'No mobile phone'})

# Create a table of counts for "answered" at each level of "mobile"
summary_table = df.groupby(['mobile', 'answered']).size().unstack()

# Create a barplot of counts of customers who answered for each level of mobile
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='answered', hue='mobile', data=df, palette="Set2")
plt.xlabel('Answered')
plt.ylabel('Count')
plt.title('Counts of Customers Who Answered by Mobile Status')
plt.legend(title='Mobile', loc='upper right')

# Adding count labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')

# Show the plot
plt.show()
```
